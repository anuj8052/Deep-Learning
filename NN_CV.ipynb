###Simple Implementation of train a neural network model using tensorflow image dataset in computer vision

```python
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
# load the cifar-10 dataset
(ds_train, ds_test), ds_info = tfds.load(
    'cifar10',
    split = ['train', 'test'],
    shuffle_files = True,
    as_supervised = True,
    with_info = True
)
```

```python
# define the preporcessing fucntion for the dataset
def preporcess(image, label):
  image = tf.cast(image, tf.float32)
  image = image/255.0
  return image, label
```

```python
# apply the preprocess fucntion to the dataset
ds_train = ds_train.map(preporcess).batch(32)
# apply the preprocess fucntion to the dataset
ds_test = ds_test.map(preporcess).batch(32)
```

```python
# define the model architecture
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu', input_shape = (32, 32, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    tf.keras.layers.MaxPooling2D((3,3)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units = 64, activation = 'relu'),
    tf.keras.layers.Dense(units = 10)
])
```

```python
# compile the model
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    metrics = ['accuracy']
)
```

```python
# training the model
model.fit(
    ds_train,
    epochs = 10,
    validation_data = ds_test
)
```

```python
loss, accuracy = model.evaluate(ds_test)
print('Test accuracy:', accuracy)
# print('Loss: ' loss)
```
